```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data/house-prices-advanced-regression-techniques/train.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
keep_cols = ["MSSubClass"
             ,"MSZoning"
             ,"LotArea"
             ,"OverallQual"
             ,"OverallCond"
             ,"YearBuilt"
             ,"CentralAir"
             ,"FullBath"
             ,"HalfBath"
             ,"BedroomAbvGr"
             ,"GarageCars"
             ,"YrSold"
             ,"MoSold"
             ,"SalePrice"
             ]

categorical_vars = ["MSSubClass"
             ,"MSZoning"
             ,"CentralAir"
             ]

target = ["SalePrice"]
```

```python
work_df = df[keep_cols]
```

```python
work_df.head()
```

```python
work_df["YearsSinceBuild"] = work_df["YrSold"] - work_df["YearBuilt"]
```

```python
work_df.drop(["YearBuilt", "YrSold"], axis=1, inplace=True)
```

```python
work_df.isna().any()
```

```python
from sklearn import preprocessing

for var in categorical_vars:
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(work_df[var])
    work_df[var] = label_encoder.transform(work_df[var])
```

```python
work_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(work_df.drop(target, axis=1), work_df[target], test_size=0.33, random_state=17)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =17)
GBoost.fit(X_train, y_train)
```

```python
preds = GBoost.predict(X_test)

diff = preds.reshape(482,1) - y_test
```

```python
diff.mean()
```
