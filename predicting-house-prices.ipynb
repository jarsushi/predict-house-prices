```python
#Print and replace all categorical field null values with 'Missing' label
def print_pct_missing_values(df, vars):
    for var in vars:
        if (np.round(df[var].isnull().mean()) > 0):
            print(var, np.round(df[var].isnull().mean(), 3),  ' % missing values')
        
print_pct_missing_values(df, categorical_vars)

for var in categorical_vars:
    df[var] = np.where(df[var].isnull(), 'Missing', df[var])
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# here I will do the model fitting and feature selection
# altogether in one line of code

# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

sel_ = SelectFromModel(Lasso(alpha=.005, random_state=17)) # remember to set the seed, the random state in this function
sel_.fit(X_train, y_train)

# this command let's us visualise those features that were kept.
# Kept features have a True indicator
sel_.get_support()

# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[(sel_.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))
print(selected_feat)
```

### Baseline Neural Network

For my baseline model, I want to create a neural network with one hidden layer the size of the input layer and capture the loss, the mean_squared_error of the exp of the data, (since the data was log transformed), and the mean error to help us set baselines to compare future models to.

```python
from keras.models import Sequential
from keras.layers import Dense

def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(units = num_features, activation='relu', kernel_initializer='normal', input_shape=(num_features,)))
    model.add(Dense(units = num_features, activation='relu'))
    model.add(Dense(1))
    # Compile model
    model.compile(loss='mean_absolute_percentage_error', optimizer='adam')
    return model
```

```python
num_features = X_train.shape[1]
model = baseline_model()
model.fit(X_train, y_train, epochs=50)
```

```python
##Mean Squared Error
predictions = np.reshape(model.predict(X_test), (len(y_test),))
# predictions = np.reshape(predictions, (438,))
exp_predictions = np.exp(predictions)
exp_y_test = np.exp(y_test)
mean_sq_err = ((exp_y_test - exp_predictions)**2).mean()

##Absolute Mean Error
abs_mean_err = abs(exp_y_test - exp_predictions).mean()

temp = {
        "Model": "baseline"
        ,"Mean Squared Error": mean_sq_err
        ,"Absolute Mean Error": abs_mean_err
       }
model_dict = {}
model_dict.update({
        "baseline": temp
})
```

### K-folds Cross Validation

Use K-folds Cross Validation for better determination of accuracy and prediction capabilities

```python
##Import libraries
from keras import models
from keras import layers
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_classification
```

```python
def create_network():
    
    # Start neural network
    network = models.Sequential()

    # Add fully connected layer with a ReLU activation function
    network.add(layers.Dense(units=(X_train.shape[1],), activation='relu', input_shape=(X_train.shape[1],)))

    # Add fully connected layer with a ReLU activation function
    network.add(layers.Dense(units=(X_train.shape[1],), activation='relu'))

    # Add fully connected layer with a sigmoid activation function
    network.add(layers.Dense(units=1, activation='sigmoid'))

    # Compile neural network
    network.compile(loss='mean_squared_error'
                    ,optimizer='adam'
                    ,metrics=['accuracy'])
    
    # Return compiled network
    return network
```

```python
# Wrap Keras model so it can be used by scikit-learn
neural_network = KerasClassifier(build_fn=create_network, 
                                 epochs=10, 
                                 batch_size=100, 
                                 verbose=0)
```

```python
# Evaluate neural network using three-fold cross-validation
cross_val_score(neural_network, X_train, y_train, cv=3)
```
